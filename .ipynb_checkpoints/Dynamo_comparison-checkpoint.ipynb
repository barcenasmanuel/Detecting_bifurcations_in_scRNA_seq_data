```python
import numpy as np
import scanpy as sc
import scvelo as scv
```

# Import adata files 

```python
filename1 = 'OVCA420_TGFB1'
adata1 = sc.read_h5ad(f'gene_exp_{filename1}.h5ad')
adata2 =sc.read_h5ad(f'dynamo_{filename1}.h5ad')
```

```python
J_gm = adata1.uns['Jacobian']['jacobians']
J_dyn = adata2.uns['mean_jacobians_dynamo']
print(f'Jacobians for our method: {J_gm.shape}')
print(f'Jacobians for dynamo method: {J_dyn.shape}')
```

# Calculate max_eig and pos eig from Dynamo mean jacobians

```python
loop = np.arange(19)
eig_list_dyn = []
max_eig_dyn, pos_dyn = [], []
for index in loop:
    w_dyn, v_dyn = np.linalg.eig(J_dyn[index])
    w_dyn = np.real(w_dyn)
    max_eig_1, pos_1 = np.max(w_dyn), np.sum(w_dyn > 0)
    eig_list_dyn.append(w_dyn)
    max_eig_dyn.append(max_eig_1)
    pos_dyn.append(pos_1)
    
eig_list_dyn = np.array(eig_list_dyn)
max_eig_dyn = np.array(max_eig_dyn)
pos_dyn = np.array(pos_dyn)
```

# Define the smooth_curve to calculate smooth pos and max_eig for our method

```python
def smooth_curve(pst, counts, intervals, rep=1):
    """
    Smooth a curve using averaging over pseudotime intervals.

    Parameters:
        pst (numpy.ndarray): Pseudotime values.
        counts (numpy.ndarray): Curve values corresponding to pseudotime.
        intervals (list): List of pseudotime intervals.
        rep (int, optional): Number of smoothing repetitions.

    Returns:
        tuple: A tuple containing smoothed pseudotime and curve values.
    """
    ref_pst, ref_counts = pst, counts
    for j in range(rep):
        avg_pst, avg_count = np.zeros(len(intervals)), np.zeros(len(intervals))

        for i in range(len(intervals)):
            t1, t2 = intervals[i][0], intervals[i][1]
            sel_pst, sel_counts = ref_pst[ref_pst > t1], ref_counts[ref_pst > t1]
            sel_pst, sel_counts = sel_pst[sel_pst < t2], sel_counts[sel_pst < t2]
            avg_pst[i], avg_count[i] = (t1 + t2) / 2., np.mean(sel_counts)
        ref_pst, ref_counts = avg_pst, avg_count

    return avg_pst, avg_count
```

# Plotting of max eig and pos eig of our method versus Dynamo

```python
import matplotlib
import matplotlib.pyplot as plt
small_size = 6
medium_size = 8
large_size = 12
tick_size = small_size
axis_size=medium_size
legend_size = small_size
sizefig = 3.54

label1 = 'Our method'
label2 = 'Dynamo'

tm = adata1.uns['Jacobian']['time']
ints = adata1.uns['Jacobian']['pst_interval']
npoints = adata1.uns['Jacobian']['npoints']

base_max_eig = adata1.uns['Jacobian']['largest']
base_pos = adata1.uns['Jacobian']['positive']

avg_max_gm, avg_pos_gm = np.mean(base_max_eig, axis=1), np.mean(base_pos, axis=1)
std_max_gm, std_pos_gm = np.std(base_max_eig, axis=1), np.std(base_pos, axis=1)

smooth_tm, smooth_avg = smooth_curve(tm, avg_max_gm, ints, rep=1)
smooth_tm, smooth_pos = smooth_curve(tm, avg_pos_gm, ints, rep=1)

matplotlib.rc('xtick', labelsize=tick_size)
matplotlib.rc('ytick', labelsize=tick_size)
fig = plt.figure(figsize=(sizefig, 2*sizefig))  # figsize=(WIDTH_SIZE,HEIGHT_SIZE); figsize = (3.54,3.54)

ax1 = plt.subplot(211)
# plt.errorbar(tm, avg_pos, yerr=std_pos, fmt='o')
plt.plot(smooth_tm, smooth_pos, 'ro-', label = label1)
plt.plot(smooth_tm, pos_dyn, 'bo-', label = label2)
plt.legend(loc='best', prop={'size': legend_size})
plt.ylabel('Positive eigenvalues', fontsize=axis_size)
plt.xlabel('Pseudotime', fontsize=axis_size)
# plt.xticks([])

ax2 = plt.subplot(212)
# plt.errorbar(tm, avg_max, yerr=std_max, fmt='o')
plt.plot(smooth_tm, smooth_avg, 'ro-', label = label1)
plt.plot(smooth_tm, max_eig_dyn, 'bo-', label = label2)
plt.legend(loc='best', prop={'size': legend_size})
plt.ylabel('Largest eigenvalue', fontsize=axis_size)
plt.xlabel('Pseudotime', fontsize=axis_size)
# plt.xticks([])

plt.tight_layout()
plt.savefig(f'Jacobian_eig_comparison_{filename1}.png', format='png', bbox_inches='tight', dpi=300)
plt.show()
```
